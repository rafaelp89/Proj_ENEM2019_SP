# Amostragem Aleatória simples
A amostragem aleatória simples é um método de seleção de amostras onde cada elemento de uma população tem a mesma chance e probabilidade conhecida de ser escolhido para fazer parte da amostra. Isso garante que a amostra seja o mais representativa possível da população, minimizando o risco de viés na seleção e permitindo que os resultados obtidos sejam generalizados com maior confiança para todo o grupo estudado. É o tipo mais básico de amostragem probabilística e serve de base para métodos mais complexos. Para o nosso projeto, por meio da amostra simples, escolhemos aleatoriamente por meio de funções Python em nosso conj de dados.

```python
#trava da escolha aleatória. Semente para fixar a escolha aleatória
np.random.seed(1)
```

```python
#Realizando a amostra simples escolhendo uma porcentagem de 20% em meu conjunto de dados extraindo 2640 linhas
amos_simples = enem_campinas.sample(frac=0.20)
```

# Amostragem Aleatória Sistemática
A amostragem aleatória sistemática é um método de seleção de uma amostra de uma população onde os elementos são escolhidos em intervalos regulares. Primeiro, a população total é organizada em uma lista e, em seguida, um ponto de partida aleatório é selecionado. A partir desse ponto, cada k-ésimo elemento é escolhido, onde k é o tamanho do intervalo de amostragem (calculado dividindo o tamanho total da população pelo tamanho da amostra desejado). Este método é simples de implementar e garante uma distribuição uniforme da amostra em toda a população. Em nosso projeto fizemos uma escolha em um intervalo de 100 em 100

```python
#tamanho dos dados de Campinas
enem_campinas.shape
```

Fazendo a escolha sistemática

```python
#Semente para fixar a escolha aleatória
np.random.seed(2)
#escolha aleatória em um intervalo
inicio = np.random.choice(10, 1)
```

```python
inicio
```

```python
#fazendo uma escolha aleatória dos índices em um internvalo de 100 em 100
amostra2 = np.arange(inicio, 13198, 100)
amostra2
```

```python
amos_sistematica = enem_campinas.iloc[amostra2, :]
```

# Amostragem Estratificada
A amostragem estratificada é uma técnica onde a população é dividida em subgrupos homogêneos, ou estratos, com base em características específicas (como idade, gênero, renda, etc.). Após essa divisão, uma amostra aleatória é selecionada de cada estrato, garantindo que todos os subgrupos importantes sejam representados proporcionalmente na amostra final. Para nosso projeto usamos a amostragem por Raça.

```python
#fazendo uma contageem dos grupos
enem_campinas['RACA'].value_counts()
```

Para cada grupo tiramos 15% de sua quantidade

```python
#extraimos 1126 pessoas do total
branca = enem_campinas.loc[enem_campinas.RACA == 'branca'].sample(frac=0.15)
branca.shape
```

```python
#extraimos 548 pessoas do total
parda = enem_campinas.loc[enem_campinas.RACA == 'parda'].sample(frac=0.15)
parda.shape
```

```python
#extraimos 42 pessoas do total
amarela = enem_campinas.loc[enem_campinas.RACA == 'amarela'].sample(frac=0.15)
amarela.shape
```

```python
#extraimos 31 pessoas do total
nd = enem_campinas.loc[enem_campinas.RACA == 'nao_declarado'].sample(frac=0.15)
nd.shape
```

```python
#extraimos 7 pessoas do total
indigena = enem_campinas.loc[enem_campinas.RACA == 'indigena'].sample(frac=0.15)
indigena.shape
```

```python
#juntar todos os grupos em uma tabela só com amostragem somente de Raças
amos_estra = pd.concat([branca, parda, amarela, nd, indigena])
amos_estra
```

# Amostragem por conglomerado ou agrupamento
A amostragem por conglomerado é uma técnica em que a população é dividida em grupos (conglomerados) heterogêneos internamente, mas homogêneos entre si. Em vez de amostrar indivíduos, seleciona-se aleatoriamente alguns desses conglomerados, e então todos os indivíduos dentro dos conglomerados escolhidos são incluídos na amostra.** É frequentemente usada quando é impraticável ou caro amostrar indivíduos diretamente, sendo comum em grandes estudos geográficos.**

```python
#localizando somente escolas públicas
escolas_publicas = enem_campinas.loc[enem_campinas.ESCOLA == 'pública']
escolas_publicas.shape
```

```python
#contagem de código das escolas
amostras = escolas_publicas['CO_ESCOLA'].value_counts()
amostras
```

```python
#extraimos os indices de cada coluna
amostras.index
```

```python
#fazemos uma escolha aleatoria de valores 0 e 1, sendo 85% de valores 0 e 15% de valores 1
amostragem = np.random.choice(a =[0,1], size=186, replace=True, p=[0.85, 0.15])
amostragem
```

```python
#relacionamos o código da escola com o valor escolhido paa a amostra 1 e extraimos somente os valores correspondentes a 1
amos_conglomerado = amostras.loc[amostragem == 1]
amos_conglomerado
```
