# Nessa parte do projeto iramos fazer a Distribuição de Frenquancias de nosso conj de dados.

```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
enem_sp = pd.read_csv('/content/drive/MyDrive/Cursos UDEMY/Curso Estatistica com Python/arquivos1/enem_2019_tratado.csv', sep=',', encoding='iso-8859-1')
```

```python
#base de dados São Paulo
enem_sp.head()
```

```python
#extraindo somente o município de Sorocaba
enem_sorocaba = enem_sp.loc[enem_sp['NO_MUNICIPIO_RESIDENCIA'] == 'Sorocaba']
enem_sorocaba.head()
```

```python
#tamanho da base de dados extraida somente de Sorocaba
enem_sorocaba.shape
```

```python
#da base de Sorocaba iremos extrair somente a coluna de idades e analizar ela
idade = enem_sorocaba['IDADE']
idade
```

```python
#analisando a frequencia absoluta de idades. Fazemos a contagem de idades. No lado esquerdo a idade e após os dois pontos no lado direito a quantidade
freq_absoluta = Counter(idade)
freq_absoluta
```

```python
#ordenando
freq_absoluta = pd.Series(freq_absoluta).sort_index(ascending=True)
freq_absoluta
```

```python
#fazendo a frequencia relativa que é a divisão do numero correspondente da frequencia absoluta de cada registro pelo soma geral da frequancia absoluta
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

```python
#percentual da frequencia relativa
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc
```
